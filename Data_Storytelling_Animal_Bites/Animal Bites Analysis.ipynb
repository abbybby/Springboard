# Does Your Pet Bite?

In the United States, animal bites are often reported to law enforcement (such as animal control). According the the World Health Organization(WHO), animal bites are a significant cause of morbidity and mortality worldwide.  
The health impacts of animal bites are dependent on the type and health of the animal species, the size and health of the bitten person, and accessibility to appropriate health care. The main concern with an animal bite is that the animal may be rabid.   
This dataset I downloaded from Kaggle includes information on over 9,000 animal bites which occurred near Louisville, Kentucky from 1985 to 2017 and includes information on the species, genders, and breeds of the animals that were involved in bite incidents and whether that animal was rabid.

## 1.Data Wrangling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('Health_AnimalBites.csv')
df.info()
```

I've downloaded this dataset from Kaggle, which includes information on over 9,000 animal bites which occurred near Louisville, Kentucky from 1985 to 2017. This dataset contains the following columns:
- bite_date: The date the bite occurred
- SpeciesIDDesc: The species of animal that did the biting
- BreedIDDesc: Breed (if known)
- GenderIDDesc: Gender (of the animal)
- color: color of the animal
- vaccination_yrs: how many years had passed since the last vaccination
- vaccination_date: the date of the last vaccination
- victim_zip: the zipcode of the victim
- AdvIssuedYNDesc: whether advice was issued
- WhereBittenIDDesc: Where on the body the victim was bitten
- quarantine_date: whether the animal was quarantined
- DispositionIDDesc: whether the animal was released from quarantine
- headsentdate: the date the animal’s head was sent to the lab
- release_date: the date the animal was released
- ResultsIDDesc: results from lab tests (for rabies)

```python
# Take a quick look at the data
df.head()
```

#### Check for Missing Values

```python
df=df.replace('UNKNOWN',np.nan)
```

```python
missing = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by=['count','%'],ascending= False)
```

```python
# remove the rows that have out-of-range bite_date (There are 10 of them)
df.drop(df[df.bite_date < '1985-01-01 00:00:00'].index,inplace=True)
df.drop(df[df.bite_date > '2017-12-31 00:00:00'].index,inplace=True)
df.shape
```

```python
# convert the bite_date to datetime objects 
df.bite_date = pd.to_datetime(df.bite_date)
# Extract year and month from bite_date and create new columns 'bite_year' and 'bite_month'
df['bite_year'] = pd.DatetimeIndex(df['bite_date']).year
df['bite_month'] = pd.DatetimeIndex(df['bite_date']).month
```

## 2.Exploratory Data Analysis

### Bites Over Time

```python
df[['bite_date','bite_year']].groupby('bite_year').count().plot(title='Number of bite cases over time', figsize=(12,6),
                                                                legend=False,xlabel='Year',ylabel='Number of bite incidents')
plt.show()
```

There is huge jump from the year of 2009 to the year of 2010. This is probably due to an increase in the recording of the events, or a change in the method of recording, rather than a sudden jump in the number of incidents.  
We can also see that the number of incidents has been steadily decreasing since the year of 2014, which is a good news.

### Number of Bites Per Month

```python
_ = df[['bite_date','bite_month']].groupby('bite_month').count().plot(title='Number of Bite Incidents Per Month', figsize=(12,6),
                                                                      kind='bar', legend=False,
                                                                      xlabel='Month',ylabel='Number of bite incidents')
plt.show()
```

As we can see from this plot, there are more incidents occurred during spring and summer months. There could be many reasons to explain this, but my assumption is that animals tend to be more aggressive when the temperature is high.

### Number of Incidents by Animal Species

```python
plt.figure(figsize=(12,6))
counts_byspecies = sns.countplot(x = 'SpeciesIDDesc', data = df, order = df.SpeciesIDDesc.value_counts().index)
for bar in counts_byspecies.patches: 
    counts_byspecies.annotate(bar.get_height(),  
                   (bar.get_x() + bar.get_width() / 2,  
                    bar.get_height()), ha='center', va='center', 
                   size=11, xytext=(0, 8), 
                   textcoords='offset points') 
plt.title('Number of Incidents Grouped by Animal Species')
plt.xlabel('Animal Species')
plt.xticks(rotation=45)
plt.show()
```

From the plot above, we can see that dog bites accounted for the majority of animal bite incidents, followed by cat, and then bat.

### Breed analysis

#### The most dangerous dog breeds

```python
dog = df[df['SpeciesIDDesc']=='DOG']
plt.figure(figsize=(12,6))
dog_species = sns.countplot(x = 'BreedIDDesc', data = dog, order = dog.BreedIDDesc.value_counts().head(10).index)
plt.title('The Most Aggressive 10 dog breeds')
plt.xlabel('Dog Breeds')
plt.ylabel('Number of Bite Incidents')
plt.xticks(rotation=45)
plt.show()
```

Pit bulls were responsible for the highest percentage of reported dog bites (15.7%), followed by German shepherd (5.6%), and Labrador Retriever (3.6%). 

#### Animal Bites and Gender(of the animal)

In general, in most mammals, males are larger than females. Is it also true that males are more aggresive than females?

```python
male_count = df[df['GenderIDDesc']=='MALE'].shape[0]
female_count = df[df['GenderIDDesc']=='FEMALE'].shape[0]
plt.pie([male_count,female_count], labels = ['Male','Female'],autopct ='%1.1f%%')
plt.title('Is Male or Female More Dangerous?')
plt.show()
```

The pie chart above reveals that most of the bite incidents were caused by male animals.  
Let's see if we can get a different result if we also consider animal species besides gender.

```python
male_dog_count = df[(df['GenderIDDesc']=='MALE')&(df['SpeciesIDDesc']=='DOG')].shape[0]
female_dog_count = df[(df['GenderIDDesc']=='FEMALE')&(df['SpeciesIDDesc']=='DOG')].shape[0]
plt.pie([male_dog_count,female_dog_count], labels = ['Male Dog','Female Dog'],autopct ='%1.1f%%')
plt.title('Are Male or Female Dogs More Dangerous?')
plt.show()
```

```python
male_cat_count = df[(df['GenderIDDesc']=='MALE')&(df['SpeciesIDDesc']=='CAT')].shape[0]
female_cat_count = df[(df['GenderIDDesc']=='FEMALE')&(df['SpeciesIDDesc']=='CAT')].shape[0]
plt.pie([male_cat_count,female_cat_count], labels = ['Male Cat','Female Cat'],autopct ='%1.1f%%')
plt.title('Are Male or Female Cats More Dangerous?')
plt.show()
```

If we just look at the incidents caused by cats, female cats were involved in 44.1% of bites, which is 10% higher than female of other species. Although male cats were still responsible for more than half of the bites, it seems like female cats are more aggressive comparing to other female mammals.

#### Where the Animal Bite

```python
plt.figure(figsize=(12,6))
counts_byspecies_body = sns.countplot(x = 'SpeciesIDDesc', data = df, order = df.SpeciesIDDesc.value_counts().index,hue='WhereBittenIDDesc')
plt.title('Where the Animal Bite? Body vs. Head')
plt.xlabel('Animal Species')
plt.xticks(rotation=45)
plt.show()
```

It's clear that people were far more likely to be bitten on their bodies than on their heads.

#### What's the probability of the animal tested positive for rabies if they bit a person?

Rabies is a viral disease that causes inflammation of the brain in humans and other mammals.  
It is usually spread through an animal bite. Animals most likely to spread rabies include dogs, bats, foxes, skunks, and raccoons.

```python
positive_count = df[df['ResultsIDDesc']=='POSITIVE'].shape[0]
negative_count = df[df['ResultsIDDesc']=='NEGATIVE'].shape[0]
plt.pie([positive_count,negative_count], explode=[0.1,0], labels = ['Positive','Negative'],colors=['r','g'],autopct ='%1.1f%%')
plt.title('Test Results for Rabies')
plt.show()
```

How likely the animal has rabies might also depends on its species.

```python
# Find the species of the animals that were tested positive for rabies
df.loc[df['ResultsIDDesc']=='POSITIVE','SpeciesIDDesc']
```

Among all reported test results, there were 299 negative results and 4 positive results. Among the 4 positive test results, 3 of them belonged to bats, and only 1 of them belonged to dogs.

```python
# See how likely a dog is rabid if it bit
dog_positive_count = df[(df['ResultsIDDesc']=='POSITIVE') & (df['SpeciesIDDesc']=='DOG')].shape[0]
dog_negative_count = df[(df['ResultsIDDesc']=='NEGATIVE') & (df['SpeciesIDDesc']=='DOG')].shape[0]
plt.pie([dog_positive_count,dog_negative_count], explode=[0.1,0], labels = ['Positive','Negative'],colors=['r','g'],autopct ='%1.1f%%')
plt.title('Dogs: Test Results for Rabies')
plt.show()
```

```python
# See how likely a bat is rabid if it bit
bat_positive_count = df[(df['ResultsIDDesc']=='POSITIVE') & (df['SpeciesIDDesc']=='BAT')].shape[0]
bat_negative_count = df[(df['ResultsIDDesc']=='NEGATIVE') & (df['SpeciesIDDesc']=='BAT')].shape[0]
plt.pie([bat_positive_count,bat_negative_count], explode=[0.1,0], labels = ['Positive','Negative'],colors=['r','g'],autopct ='%1.1f%%')
plt.title('Bats: Test Results for Rabies')
plt.show()
```

From the charts above, we can conclude that there's approximately 1.3% of chance that the animal has rabies if it bit.
